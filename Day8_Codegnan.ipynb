```python
 # List Method
 '''
    - append(ele):- it adds the element at end of the list
    - extend(list_eles):- It adds all list elements at end of the list  
    - insert(pos,val):- It inserts the value at particular position
    '''
    
# removing element from list
''' 
    - remove(val):- It removes the first occurance value if value is present in list, otherwise valueError
    - pop():-  It removes and returns the last element in the list
    - pop(pos):- It remove and returns specified position element from the list.  '''

#  index finding for an element

''' 
    - index(val):- It return the first occurance value index position '''

#  sorting list

'''  - sort():- It sorts the original list into asscending order
     - sort(reverse = True):- It sorts the original list into descanding order '''


```

```python
shiva = [2,4,6,8,10]
shiva.append(100)
print(shiva)
shiva.extend([12,14,16])  # Extend add with bulk data [1,2,3,4]  
print(shiva)
shiva.insert(0,"shivaji")
print(shiva)
shiva.remove("shivaji")
print(shiva)
shiva.pop()
print(shiva)
lst = [1,6,2,7,3,4,8,5]
lst.sort()
print(lst)
lst.clear()
print(lst)
s = [2, 4, 6, 2,8, 10,2, 100]
ind = s.index(4)
print(ind)
print(s.count(2))
```

```python
# Buliting functions
#  len(), min(), max(), sum()

lst = [ 10,20,30,40]
print(min(lst), len(lst), max(lst), sum(lst))
```

```python
#  deep copy
l1 = [1,2,3]
l2 = [1,2,3,]
print(l1 == l2)
print(l1 is l2)
```

```python
#  shallow copy
l1 = [1,2,3]
l2 = l1
print(l1 == l2)
print(l1 is l2)
l1.append(10)
print(l1)
```
