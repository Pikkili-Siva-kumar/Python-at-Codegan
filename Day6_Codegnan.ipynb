```python
shiva = "shivakumar"
print(shiva[0:5])     # shiva
print(shiva[5:10])    # kumar
print(shiva[:5])      # shiva
print(shiva[5:])      # kumar
print(shiva[:])       # shivakumar
print(shiva[3:9])      # vakuma
print(shiva[3:9:2])    # vkm
print(shiva[-9:-3])    # ivakum
print(shiva[-1:-6:-1]) # ramuk
print(shiva[::-1])     # ramukavihs
print(shiva[::])       # shivakumar
print(shiva[1::3])     # hakr
print(shiva[-1::-2])   # rmkvhs 
print(shiva[0:10:2])  # sivkmr
print(shiva[1:10:2])  # hakua
print(shiva[::2])     # sivkmr
print(shiva[::3])     # skmr
print(shiva[::-1])    # ramukavihs
print(shiva[::-2])    # rmkvhs
print(shiva[-5:])     # kumar
print(shiva[-10:-5])  # shiva
print(shiva[-7:-2])   # vakum
print(shiva[2:-2])    # ivakum
print(shiva[-8:8])    # ivakuma
print(shiva[1:-1])    # hivakuma
print(shiva[-1:-5:-1])   # ramu
print(shiva[-3:-8:-1])   # mukavi
print(shiva[8:2:-1])     # mukavi
```
