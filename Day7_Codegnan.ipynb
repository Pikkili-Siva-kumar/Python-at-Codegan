```python
# String methods:
''' - Case conversin methods
    - upper()
    - lower()
    - title()
    - swapcase()
    - capitalize() '''


s = " The sky is blue and Ocean is BLUE"
print(s.upper())           # It returns upper  case converted string
print(s.lower())           # It converts entire string into lower case 
print(s.title())           # It converts the first character in each word into uppercase and remining all are into lower case
print(s.swapcase())        # It converts the lowercase character into uppercase and vise versa
print(s.capitalize())      # It converts first character of string into uppercase and remining all are intlo lower case
```

```python
# white space or pattern removing from string

''' 
   - strip(char)  : Return a copy of the string with leading and trailing whitespace removed. 
   - lstrip()     : Return a copy of the string with leading whitespace removed.
   - rstrip()     : Return a copy of the string with trailing whitespace removed. 
   
   '''

name = "  shiva kumar  "
print(name.strip())
print(name.lstrip())
print(name.rstrip())
```

```python
name = "@@@@@ shiva kumar@@"
print(name)
print(name.strip('@'))
print(name.lstrip('@'))
print(name.rstrip('@'))
```

```python
# string checking method

'''
   - islower() :- it returns ture of entire string is lower
   - isupper() :- it returns ture of entire string is upper
   - istitle() :- it returns ture of entire string is title manner
   - islpha() :- it returns ture of entire string contains only alphabets
   - isnum() :- it returns ture of entire string contain only digits and alphabets
   - isdigit():- it returns ture of entire string contains only digits
   - startswith(sun_string) :- it returns ture of entire string starts sub_string
   - endswith(sub_subtring) :- it returns ture of entire string ends sub_string    '''

s = "shiva"
print(s.isupper())
print(s.islower())
print(s.istitle())
print(s.isalpha())
print("shiva123".isdigit())
print("shivakumar247".isalnum())
print(s.startswith('s'))
print(s.endswith('r'))
```
